```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
for project_path in[
                    Path("BINARY_FOLD_622_label"),
                    Path("BINARY_FOLD_811_label"),
                    Path('MULTI_FOLD_622_sev_cars'),
                    Path('MULTI_FOLD_811_sev_cars'),
                    Path('MULTI_FOLD_811_old_cars'),
                    Path('MULTI_FOLD_811_young_cars'),                           
                    ]    :
    for data_type in ['IJA', 'RJA_LOW', 'RJA_HIGH'] : 
        df_path = project_path.joinpath(data_type, 'participant_information_df.csv')
        df = pd.read_csv(df_path)
        df['is_used_for_test'] = (df.filter(like = 'fold') == 2).sum(axis = 1)
        unique_participant = df.query('is_used_for_test == False')['id'].unique()
        print(f"{project_path.as_posix(): <25} {data_type: <20}, {len(unique_participant)}")
```
